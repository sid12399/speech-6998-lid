```python
from glob import glob


root_dir = "/home/sp4013/kaldi-trunk/egs/lid/s1/db/cu-multilang-dataset/"

for fol in glob(root_dir + "*/"):
    utt2spk = []
    with open(fol + "wav.scp") as infile:
        for line in infile.readlines():
            utt_id = line.split()[0]
            speaker_name = utt_id[:-6]
            utt2spk.append((speaker_name, utt_id))
            
    with open(fol + "utt2spk", "w") as outfile:
        for spk, utt in utt2spk:
            outfile.write(spk + "    " + utt + "\n")
            
print("Done")
```
